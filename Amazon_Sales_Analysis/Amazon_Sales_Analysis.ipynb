# üìä Amazon-Style Sales Data Analysis using Python  

**Objective:**  
To analyze sales, profit, and regional performance using Python and derive business insights.  

**Tools Used:**  
Pandas | Matplotlib | Seaborn  

**Dataset Description:**  
Contains over 3,200 sales transactions with:  
- Order details (Order ID, Order Date, Ship Date)  
- Customer location (Geography)  
- Product and category information  
- Key metrics: Sales, Quantity, Profit  

**Goal:**  
To uncover sales trends, top-performing products, and regions driving maximum profitability.


## üì• Step 1: Importing the Dataset  
We start by uploading the Excel dataset to Google Colab and reading it using Pandas.


```python
from google.colab import files
uploaded = files.upload()
```

## üîç Step 2: Loading and Previewing the Dataset  
After uploading, we load the Excel file into a Pandas DataFrame and preview the first few records.


```python
import pandas as pd

# Load the dataset
df = pd.read_excel("Amazon Sales Dataset.xlsx")
df.head()
```

## üìä Step 3: Exploring the Dataset  
We check for data types, missing values, duplicates, and overall data structure.


```python
# Basic overview
df.info()

# Summary statistics
print("\nSummary Statistics:")
display(df.describe())

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Check duplicates
print("\nDuplicate Records:", df.duplicated().sum())
```

## üìà Step 4: Visualizing Sales and Profit Trends  
Visualizing sales and profit distribution by category, geography, and product performance.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sales by Category
plt.figure(figsize=(8,4))
sns.barplot(x='Category', y='Sales', data=df, estimator='sum', palette='crest')
plt.title("Total Sales by Category")
plt.xticks(rotation=45)
plt.show()

# Profit by Category
plt.figure(figsize=(8,4))
sns.barplot(x='Category', y='Profit', data=df, estimator='sum', palette='flare')
plt.title("Total Profit by Category")
plt.xticks(rotation=45)
plt.show()

# Top 10 Products by Sales
top_products = df.groupby('Product Name')['Sales'].sum().nlargest(10)
top_products.plot(kind='bar', figsize=(10,4), title='Top 10 Products by Sales', color='teal')
plt.ylabel("Total Sales")
plt.show()
```

## üåç Step 5: Regional Sales Performance  
Analyzing which geographical regions generated the highest total sales.


```python
top_geo = df.groupby('Geography')['Sales'].sum().nlargest(10)
top_geo.plot(kind='barh', figsize=(8,5), title='Top 10 Locations by Sales', color='skyblue')
plt.xlabel("Total Sales")
plt.show()
```

## üí∞ Step 6: Key Performance Indicators (KPIs)  
Summarizing major business metrics from the dataset.


```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
avg_profit_margin = (total_profit / total_sales) * 100

print(f"üí∞ Total Sales: ${total_sales:,.2f}")
print(f"üèÜ Total Profit: ${total_profit:,.2f}")
print(f"üì¶ Total Orders: {total_orders}")
print(f"üìä Average Profit Margin: {avg_profit_margin:.2f}%")
```

## üß† Step 7: Insights  

- Los Angeles generated the **highest total sales** among all locations.  
- The **‚ÄúFurnishings‚Äù** category contributed significantly to total revenue.  
- The **average profit margin** stands around **18‚Äì20%**, indicating strong profitability.  
- Sales show steady growth over multiple years, highlighting consistent business performance.
